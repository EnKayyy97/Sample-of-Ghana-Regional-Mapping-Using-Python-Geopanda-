```python
#Customizing Plots
myRegions.plot(figsize=(10,6), edgecolor='k', cmap='Pastel2')
plt.title("Ghana")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(False)
plt.show()
```

```python
#Plotting Ghana Regions and Capitals
ax = myRegions.plot(figsize=(10, 5), edgecolor='k', cmap='Pastel1')
regionscentroid.plot(ax=ax, color='red', markersize=7)
plt.title("Ghana Regions with Their Capital")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
